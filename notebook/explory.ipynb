```python
import pandas as pd 
import numpy as np
from tensorflow.keras.utils import to_categorical
```

```python
train = pd.read_csv('../input/Train.csv')
train.head()
```

```python
train.head()
```

```python
train["label"] = train["type"].astype('category').cat.codes
```

```python
train.head()
```

```python
labels_ord = ['Harmful_Traditional_practice','Physical_violence', 'economic_violence', 'emotional_violence','sexual_violence']
```

```python
Y = to_categorical(train['label'])
```

```python
for i in range(len(labels_ord)) :

    train[labels_ord[i]] = Y[:,i]
```

```python
train.head()
```

```python
train.to_csv("../input/Mytrain.csv" , index=False)
```
